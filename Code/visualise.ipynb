```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

```

```python
# List of your 9  file paths
csv_files = [
    'heart_RandomForest', 'diabetes_RandomForest', 'fraud_RandomForest',
    'heart_MLP', 'diabetes_MLP', 'fraud_MLP', 
    'heart_LR', 'diabetes_LR', 'fraud_LR'
]
```

```python
# Manually specified values to be highlighted for each plot
highlight_values = [0.740, 0.7467, 0.91,
                    0.7078, 0.6818, 0.935,
                    0.7468, 0.7467, 0.935]
```

```python
fig, axes = plt.subplots(3, 3, figsize=(15, 12))  # 3x3 grid of subplots
axes = axes.flatten()

for i, (file, highlight) in enumerate(zip(csv_files, highlight_values)):
    # Load CSV
    df = pd.read_csv(file + ".csv", sep = ';', header= 0)

    # Plot on its corresponding axis
    sns.boxplot(y='value', data=df, ax=axes[i])

    # Add the horizontal line for the specific value
    axes[i].axhline(highlight, color='red', linestyle='--', label=f'Base model accuracy: {highlight}')
    axes[i].set_title(f'{csv_files[i]}')
    axes[i].legend()
plt.savefig('boxplots.png', dpi=300)  # Save the figure
plt.tight_layout()
plt.show()

```
